## Medium complexity - Birthday Paradox

Birthday paradox concerns the probability that, in a set of n randomly chosen people, some pair of them will have the same birthday. You can read about this problem at https://en.wikipedia.org/wiki/Birthday_problem.

Write a function called <i><b>has_duplicates</b></i>, that takes a list and returns <i>True</i> if there is any element that appears more than once. It should not modify the original list.

If there are 45 students in the class, what are the chances that two of you have the same birthday? You can estimate this probability by generating random samples of 45 birthdays and checking for matches.

P(A) = Probalility of atleast 2 people having the same birthday in class
P(A') = 1- P(A) = Probability that none of the students in class have same birthday

To calculate P(A') ,
the 1st student can have any birthdate i.e. option of picking 365 days from 365 days.
2nd student can now have any birthdate expect 1st student's i.e. option of picking 364 days from 365 days
3rd student has 363 options from 365 days and so on..

P(A') = 365/365 * 364/365 * 363/365 * 362/365 * ......n times 

P(A) = 1 - (365/365 * 364/365 * 363/365 * 362/365 * ......n times)

```python
# Theoritical Calculation

def prob_theoritical(n):
    prob = 1
    for index in range (0,n):
        aProb = (365-index)/365
        prob = prob*aProb
    return 1- prob
        
```

```python
# For Class of 45 students, the probality of having atleast 2 student sharing same birthday is

print ("Probality for 45 students: " , prob_theoritical(45))
```

```python
# function to find if there is any duplicate in the list
# function return 'True' if duplicate exist else returns 'False'

# input : list
# output : boolean

def has_duplicates (l):
    s = l[:] # create copy of original list 'l'
    s.sort()
    
    for index in range(0,len(s)-1):
        if (s[index]==s[index+1]):
            return True
    else:
        return False   

#------------------------------------------------------------------------------------------------------------
    

# This function creates a list of random birthdays of size 'n' 
# assumption : all the birthdays are from non-leap year
# i.e. there are 365 days in an year

# input : length of the list to generate
# output : list

import random

def generate_birthday_list(n):
    s = []
    for i in range(0,n):
        birthday = random.randint(1,365)
        s.append(birthday)
    return s

#------------------------------------------------------------------------------------------------------------


# Calculate probability of any two students having the same birthday 
# in a class of size 'n' for sample size 'k' 

# input : Class size (n) and Sample Size (k)

def calculate_prob (n,k,enablePrint):
    found_duplicate=0
    for i in range (0,k):
        s = generate_birthday_list(n)
        if (enablePrint):
            print('\033[34m' + "Sample" , i+1 , " : " , s, '\033[35m' )
        if (has_duplicates(s)):
            if (enablePrint):
                print("")
                print ('\033[32m' + "Duplicates list: " ,find_duplicates(s), '\033[33m' )
                print ("-----------------------------------------------------------------------------------")
            found_duplicate+=1
        else:
            if (enablePrint):
                print("------------------------------------------------------------------------------------")
    prob = found_duplicate/k
    return prob


#------------------------------------------------------------------------------------------------------------

# Find duplicates in a list
# input : list
# out : list of duplicates

def find_duplicates (l):
    duplicate_list = []
    s = l[:] # create copy of original list 'l'
    s.sort()
    index = 0
    while (index < len(s)-1):
        if (s[index]==s[index+1]):
            #duplicate value found
            dup_value = s[index]
            duplicate_list.append(s[index])
            #find additional duplicates of dup_value
            while (index+1 < len(s)):
                if (s[index+1]==dup_value):
                    duplicate_list.append(s[index+1])
                    index+=1
                else:
                    index+=1
                    break
        else:
            index+=1
    return duplicate_list

#------------------------------------------------------------------------------------------------------------


# MAIN FUCNTION TO PRINT FOLLOWING

"""
– number of students,
– number of iterations/samples
– list of duplicate days for each iteration, where
duplicates are found.
"""
# Probality here is the probability of having any 2 students having the same birthday in the class

# input : Class size (n) and Sample Size (k)

def main(n,k,enablePrint):
    
    print ('\033[32m' + "Number of students in class : ", n , '\033[33m')
    print ('\033[32m' + "Number of iterations/samples used: ", k, '\033[33m')
    print ('\033[32m' + "------------------------------------------------------------------------------------" + '\033[33m')
    print ('\033[31m' + "Probability: ", calculate_prob(n,k,enablePrint), '\033[32m')
```
